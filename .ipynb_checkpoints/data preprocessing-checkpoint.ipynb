```python
import os
import cv2
import matplotlib.pyplot as plt
```

```python
face_cascade = cv2.CascadeClassifier('haarcascades/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('haarcascades/haarcascade_eye.xml')
```

```python
def get_cropped_image_if_2_eyes(image_path):
    try:
        img = cv2.imread(image_path)
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, 1.3, 5)
        for (x,y,w,h) in faces:
            roi_gray = gray[y:y+h, x:x+w]
            roi_color = img[y:y+h, x:x+w]
            eyes = eye_cascade.detectMultiScale(roi_gray)
            if len(eyes) >= 2:
                return roi_color
    except Exception as e:
        pass  
```

```python
path_to_data = "data"
path_to_cr_data = "data/cropped/"
```

```python
import os
img_dirs = []
for entry in os.scandir(path_to_data):
    if entry.is_dir():
        img_dirs.append(entry.path)
```

```python
img_dirs
```

```python
import shutil
if os.path.exists(path_to_cr_data):
     shutil.rmtree(path_to_cr_data)
os.mkdir(path_to_cr_data)
```

```python
cropped_image_dirs = []
celebrity_file_names_dict = {}

for img_dir in img_dirs:
    count = 1
    celebrity_name = img_dir.split('/')[-1]
    print(celebrity_name)
    
    celebrity_file_names_dict[celebrity_name] = []
    
    for entry in os.scandir(img_dir):
        roi_color = get_cropped_image_if_2_eyes(entry.path)
        if roi_color is not None:
            cropped_folder = path_to_cr_data + celebrity_name
            if not os.path.exists(cropped_folder):
                os.makedirs(cropped_folder)
                cropped_image_dirs.append(cropped_folder)
                print("Generating cropped images in folder: ",cropped_folder)
                
            cropped_file_name = celebrity_name + str(count) + ".png"
            cropped_file_path = cropped_folder + "/" + cropped_file_name 
            
            cv2.imwrite(cropped_file_path, roi_color)
            celebrity_file_names_dict[celebrity_name].append(cropped_file_path)
            count += 1    
```

```python
def get_cropped_faces(img_path):
    cropped = []
    img = cv2.imread(img_path)
    faces = face_cascade.detectMultiScale(img, 1.3, 5)
    for face in faces:
        x,y,w,h = face
        cropped_face = img[y:y+h,x:x+w]
        eyes = eye_cascade.detectMultiScale(cropped_face)
        if len(eyes) >= 2:
            cropped.append(cropped_face)
    return cropped
```

```python
raw_path = 'data'
cropped_path = 'data/cropped/'
```

```python
def raw_to_cropped(raw_path, cropped_path):
    if os.path.exists(cropped_path):
        os.rmdir(cropped_path)
    for i in os.scandir(raw_path):
        class_name = i.name
        class_path = i.path 
        cropped_class_path = os.path.join(cropped_path, class_name)
        img_num=0
        for j in os.scandir(class_path):
            img_path = j.path
            cropped_faces = get_cropped_faces(img_path)
            for face in cropped_faces:
                file_name = class_name + str(img_num) + '.jpg'
                file_path = os.path.join(cropped_path, file_name)
                cv2.imwrite(file_path, face)
                img_num += 1
            
```
